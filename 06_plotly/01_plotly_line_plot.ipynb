```python
# İlk 100 üniversiteyi seçip df_top100 isminde yeni bir df yaratalım
df_top100 = df.iloc[:100,:]
```

```python
df_top100.head()
```

```python
# 1. çizgi
trace1 = go.Scatter(
    x = df_top100.world_rank,
    y = df_top100.citations,
    mode = "lines",
    name = "citations",
    marker = dict(color = 'rgba(16,112,2,0.5)'), # ilk üç rakam renk, son rakam saydamlık
    text = df_top100.university_name # # hover durumunda görünecek bilgi
)
# https://www.rapidtables.com/web/color/RGB_Color.html renkleri seçmemize yardımcı olabilir.
```

```python
# 2. çizgi
trace2 = go.Scatter(
    x = df_top100.world_rank,
    y = df_top100.teaching,
    mode = "lines+markers",
    name = "teaching",
    marker = dict(color = 'rgba(80, 26, 80, 0.5)'), # ilk üç rakam renk, son rakam saydamlık
    text = df_top100.university_name # hover durumunda görünecek bilgi
)
# https://www.rapidtables.com/web/color/RGB_Color.html renkleri seçmemize yardımcı olabilir.
```

```python
data = [trace1, trace2]
```

```python
layout = dict(title = 'Teaching Citation Karşılaştırma',
             xaxis= dict(title="World Rank", ticklen=5, zeroline=False))
```

```python
fig = dict(data=data, layout=layout)
```
