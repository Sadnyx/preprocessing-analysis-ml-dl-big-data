```python
class nesne:
    def __init__(self, sayılistesi):
        self.liste = sayılistesi
   
    def toplam(self):
        sum = 0
        for i in self.liste:
            sum += i
        return sum

    def ortalama(self):
        return self.toplam() / len(self.liste)

    def standart_sapma(self):
        dummy = 0
        for i in self.liste:
            dummy += (i - self.ortalama())**2
        return (dummy / (len(self.liste)-1))**0.5
```

```python
nesne = nesne([1,2,3,4,5])
```
