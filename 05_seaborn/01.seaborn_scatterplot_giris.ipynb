    Seaborn matplotlib'i temel alan bir Python veri görselleştirme kütüphanesidir. Çekici ve bilgilendirici istatistiksel grafikler çizmek için üst düzey bir arayüz sağlar.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
```

### Saçılma Diyagramı için en temel kural iki değişkenin de nümerik olmasıdır

```python
bahsis_df = sns.load_dataset("tips")
bahsis_df.head()
```

```python
sns.relplot(x="total_bill", y="tip", kind="scatter", data=bahsis_df)
```

```python
# seaborn dataframe içindeki niteliklerin ne olduğunu bilir.
# dolayısıyla x ve y değişkenlerine nitelik/sütun isimlerini vermemiz yeterlidir.
```

```python
# aynı grafik scatterplot() metoduyla da çizilir.
```

```python
sns.scatterplot(x="total_bill", y="tip", data=bahsis_df)
```

## Kategorik değişkenlerle grafiğin çok boyutlu hale getirilmesi (semantic variable )

### Renk değişikliği ile hue

```python
# bahşiş verenlerden sigara içen içmeyen durumunu grafiğe ekleyelim.
# Nokta sayısı ve şekli değişmiyor ancak renkler içen içmeyen diye ayrılıyor.
# hue ile smoker niteliğini eklediğimizde smoker kategorileri (Yes,No) farklı renkleniyor.
```

```python
sns.relplot(x="total_bill", y="tip", hue="smoker", kind="scatter", data=bahsis_df)
```

### Şekil değişikliği ile style

```python
# style parametresine time dediğimizde time kategorilerden Lunch ve Dinner şekil değiştirerek grafikte yer alır
```

```python
sns.relplot(x="total_bill", y="tip", hue="smoker", style="time", kind="scatter", data=bahsis_df)
```

### Büyüklük değişikliği ile size

```python
sns.relplot(x="total_bill", y="tip", size="size", sizes=(15, 200), kind="scatter", data=bahsis_df);
```
